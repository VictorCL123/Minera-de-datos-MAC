### Data Transform 

```python
df_dropped = df.drop(columns=['pickup_community_area'])
df_dropped = df.dropna(subset=['taxi_id','trip_total'])
df_dropped.head()
```

```python
df.shape
```

```python
df_dropped.shape
```

```python
df_dropped['trip_total'].dtype
```

```python
df_dropped['trip_total'].mean()
```

```python
df_copy = df.copy()
```

```python
df_copy.shape
```

```python
df_copy['trip_total'].fillna(value=np.round(df['trip_total'].mean(),decimals=2),inplace = True)
```

```python
df_copy['payment_type'].unique()
```

```python
df['payment_type'].replace({'Cash':1,
                 'Credit Card':2,
                 'Unknown':0,
                 'No Charge':0,
                 'Pcard':2,
                 'Prcard':2,
                 'Dispute':0},inplace=True)
```
