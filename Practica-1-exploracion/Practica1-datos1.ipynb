## Data Transform

```python
df_dropped = df.dropna()
df_dropped.head()
```

```python
df_dropped.shape
```

```python
df_copy = df.copy()
df_copy.head()
```

```python
df_copy.shape
```

```python
df_dropped['testelapse'].dtype
```

```python
df_dropped['testelapse'].mean()
```

```python
df_copy['EXT1'].value_counts(dropna = False)
```

```python
var_disc = list(df.columns)
```

```python
var_disc = [x for x in var_disc if len(x)<5]
```

```python
for i in var_disc:
    df_copy[i].fillna(value=np.round(df[i].mean(),decimals=0),inplace=True)
```

```python
list(df['EXT1'].unique())
```

```python
for i in var_disc:
    df[i].replace({1:"Strongly Disagree",
                    2:"Disagree",
                    3:"Neutral",
                    4:"Agree",
                    5:"Strongly Agree"},inplace=True)
```
